# Data Cleaning

## Checking for any missing values in the data

```python
data.isnull().sum()
```

```python
# No missing values found
```

```python
data[cat_cols].nunique()
```

```python
# all the columns that have 6 unique values sholud have only 5 uniques instead corresponding to each of the rating
# values [1, 5], but the the sixth value which is 0 corresponds to an unrated service, so a good representative imputation
# would be to replace these 0 values of each column with the mode value of that particular column.
```

```python
impute_cols = [val for val, boolean in zip(cat_cols, list(data[cat_cols].nunique() == 6)) if boolean]
impute_cols
```

```python
modes = data[impute_cols].mode()
data[impute_cols] = data[impute_cols].replace({'0':modes})
test_data[impute_cols] = test_data[impute_cols].replace({'0':modes})
```

## Feature Transformation

```python
# Flight Distance and Departure Delay in Minutes are skewed, log transformation should be applied to them.

skewed_cols = ['Flight Distance', 'Departure Delay in Minutes']
data[skewed_cols] = np.log1p(data[skewed_cols])
test_data[skewed_cols] = np.log1p(test_data[skewed_cols])

for col in skewed_cols:
    fig, ax = plt.subplots(1, 2, figsize = (20, 5))
    sns.kdeplot(data[col], ax = ax[0])
    sns.boxplot(x = data[col], ax= ax[1])
```

## Encoding

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
le_cols = ['Gender', 'Customer Type', 'Type of Travel', 'satisfaction']

for col in le_cols:
    data[col] = le.fit_transform(data[col])
    test_data[col] = le.transform(test_data[col])

data = pd.get_dummies(data, columns=['Class'])
test_data = pd.get_dummies(test_data, columns=['Class'])

data
```

## Feature Scaling

```python
X_train = data.drop('satisfaction', axis=1)
y_train = data.satisfaction

X_test = test_data.drop('satisfaction', axis=1)
y_test = test_data.satisfaction
```

```python
from sklearn.preprocessing import RobustScaler

scaler = RobustScaler()
scaler.fit(X_train)

X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

# Modelling using XGBoost

```python
import xgboost as xgb
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay

model = xgb.XGBClassifier(use_label_encoder=False)

model.fit(X_train_scaled, y_train)

y_pred = model.predict(X_test_scaled)

print(classification_report(y_test, y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred, labels=model.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_)
disp.plot()
plt.show()
```
